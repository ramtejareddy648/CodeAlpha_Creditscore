```python
import pandas as pd
from sklearn.utils import resample

# Assume your original data is in df
# And 'label' column is your target

# Separate each class
df_class_0 = df[df['Credit Score'] == 0]
df_class_1 = df[df['Credit Score'] == 1]
df_class_2 = df[df['Credit Score'] == 2]

# Upsample class 0 and 2 to 75 samples
df_class_0_upsampled = resample(df_class_0,
                                replace=True,
                                n_samples=75,
                                random_state=42)

df_class_2_upsampled = resample(df_class_2,
                                replace=True,
                                n_samples=75,
                                random_state=42)

# Combine all classes
df_balanced = pd.concat([df_class_0_upsampled, df_class_1, df_class_2_upsampled])

# Shuffle the dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Split into X and y
x= df_balanced.drop('Credit Score', axis=1)
y = df_balanced['Credit Score']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB


```

```python
models={
    'knn':KNeighborsClassifier(),
    'decision':DecisionTreeClassifier(),
    'random':RandomForestClassifier(),
    'gradient':GradientBoostingClassifier(),
    'naive':MultinomialNB()
    
}
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    pipe=make_pipeline(model)
    pipe.fit(x_train,y_train)
    
    y_train_pred=pipe.predict(x_train)
    y_test_pred=pipe.predict(x_test)
    
    
    print(list(models.keys())[i])
    print('x_train prediction ')
    print('accuracy',accuracy_score(y_train,y_train_pred))
    print('classification_report',classification_report(y_train,y_train_pred))
    print('--'*32)
    print('x_test prediction ')
    print('accuracy',accuracy_score(y_test,y_test_pred))
    print('classification_report',classification_report(y_test,y_test_pred))
    print('----'*23)
```
